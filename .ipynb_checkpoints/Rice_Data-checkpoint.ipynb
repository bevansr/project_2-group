```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
```

```python
file_landmarks = "Resources/Landmarks%20176-318%20-%20Sheet1.csv"
file_murals = "Resources/Mural_Registry.csv"

landmarks_df = pd.read_csv(file_landmarks)
murals_df = pd.read_csv(file_murals)
```

```python
landmarks_df
```

```python
landmarks_df.isna().sum()
```

```python
landmarks_df.columns= landmarks_df.columns.str.lower()
landmarks_df
```

```python
landmarks_df = landmarks_df.rename(columns={'landmark name': 'name', 'architect': 'artist', 'date built': 'year installed'})
landmarks_df
```

```python
landmarks_df.columns
```

```python
landmarks_df = landmarks_df[['name', 'address', 'year installed', 'artist',
       'landmark designation date', 'latitude', 'longitude','zip code', 'type']]
landmarks_df
```

```python
#landmarks_df = landmarks_df[landmarks_df['latitude'].dropna()]
#landmarks_df = landmarks_df[landmarks_df['longitude'].dropna()]
#landmarks_df = landmarks_df[landmarks_df['landmark designation date'].notna()]


landmarks_df.dropna(subset=['longitude', 'latitude'])
landmarks_df['year installed'].fillna(0, inplace=True)
landmarks_df
```

```python
murals_df
```

```python
murals_df = murals_df.rename(columns={'Mural Registration ID': 'ID','Artist Credit': 'Artist', 'Zip': 'Zip Code', 'Street Address': 'Address', 'Artwork Title': 'Name'})
murals_df
```

```python
murals_df.columns
```

```python
murals_df = murals_df[['Artist', 'Name',
       'Year Installed', 'Address', 'Zip Code', 'Description of Artwork',
       'Latitude', 'Longitude', 'Type']]
murals_df
```

```python
#murals_df = murals_df[murals_df['Latitude'].notna()]
#murals_df = murals_df[murals_df['Longitude'].notna()]
murals_df = murals_df[murals_df['Zip Code'].notna()]
murals_df = murals_df[murals_df['Year Installed'].notna()]
murals_df.dropna(subset=['Longitude', 'Latitude'])
murals_df
```

```python
murals_df.columns= murals_df.columns.str.lower()
murals_df
```

```python
murals_df['year installed'].fillna(0, inplace=True)
murals_df
```

```python
landmarks_df.dtypes
```

```python
murals_df.dtypes
```

```python
murals_df['zip code'] = murals_df['zip code'].astype(int)
murals_df['year installed'] = murals_df['year installed'].astype(int)
murals_df['longitude'] = pd.to_numeric(murals_df['longitude'], errors='coerce')


# df['DataFrame Column'] = pd.to_numeric(df['DataFrame Column'], errors='coerce')
```

```python
murals_df.dtypes
```

```python
landmarks_df.dtypes
```

```python
chicago_df = pd.merge(landmarks_df, murals_df, on=['name', 'artist', 'year installed', 'address', 'latitude', 'longitude', 'type', 'zip code'], how='outer')
chicago_df
```

```python
chicago_df.columns = chicago_df.columns.str.replace(' ', '_')
chicago_df
```

```python
chicago_df['zip_code'] = chicago_df['zip_code'].astype(str)

chicago_df
```

```python
chicago_df['name'] = chicago_df['name'].replace(np.nan, 'Unknown', regex=True)
chicago_df['address'] = chicago_df['address'].replace(np.nan, 'Unknown', regex=True)
chicago_df['artist'] = chicago_df['artist'].replace(np.nan, 'Unknown', regex=True)
chicago_df['landmark_designation_date'] = chicago_df['landmark_designation_date'].replace(np.nan, 'Unknown', regex=True)
chicago_df['zip_code'] = chicago_df['zip_code'].replace(np.nan, 'Unknown', regex=True)
chicago_df['description_of_artwork'] = chicago_df['description_of_artwork'].replace(np.nan, 'None', regex=True)
murals_df.dropna(subset=['longitude', 'latitude'])
chicago_df
```

```python
# THIS IS MY TROUBLESHOOTING CELL
#chicago_df = chicago_df[chicago_df['latitude'].notna()]
#chicago_df = chicago_df[chicago_df['longitude'].notna()]
#TS = chicago_df.loc[chicago_df['name'] == 'We Grow Here Too', :]
#TS
```

```python
chicago_df = chicago_df.replace(',', ';', regex=True)
chicago_df = chicago_df.replace('�', ';', regex=True)
chicago_df = chicago_df.replace('\n', ' ', regex=True)
chicago_df
```

```python
chicago_df.dtypes
```

```python
chicago_df['longitude'] = chicago_df['longitude'].astype(float)
```

```python
chicago_df.dtypes
```

```python
chicago_df.columns
```

```python
engine = create_engine(f"postgresql://postgres:6472@localhost:5432/Landmarks")

engine.table_names()
```

```python
chicago_df.to_sql(name='chicago_df', con=engine, if_exists='append', index=False)
```

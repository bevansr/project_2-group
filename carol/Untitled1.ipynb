```python
%matplotlib inline
from sqlalchemy import create_engine
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_landmarks = "Resources/AllLandmarks.csv"
file_murals = "Resources/Mural_Registry.csv"

landmarks_df = pd.read_csv(file_landmarks)
murals_df = pd.read_csv(file_murals)
```

```python
landmarks_df
```

```python
landmarks_df.columns= landmarks_df.columns.str.lower()
landmarks_df
```

```python
landmarks_df = landmarks_df.rename(columns={'landmark name': 'name', 'architect': 'artist', 'date built': 'year installed'})
landmarks_df
```

```python
landmarks_df.columns
```

```python
landmarks_df = landmarks_df[['name', 'address', 'year installed', 'artist',
       'landmark designation date', 'latitude', 'longitude','zip code', 'type']]
landmarks_df
```

```python
landmarks_df = landmarks_df[landmarks_df['latitude'].notna()]
landmarks_df = landmarks_df[landmarks_df['longitude'].notna()]
landmarks_df
```

```python
murals_df
```

```python
murals_df = murals_df.rename(columns={'Mural Registration ID': 'ID','Artist Credit': 'Artist', 'Zip': 'Zip Code', 'Street Address': 'Address', 'Artwork Title': 'Name'})
murals_df
```

```python
murals_df.columns
```

```python
murals_df = murals_df[['Artist', 'Name',
       'Year Installed', 'Address', 'Zip Code', 'Description of Artwork',
       'Latitude', 'Longitude', 'Type']]
murals_df
```

```python
murals_df = murals_df[murals_df['Latitude'].notna()]
murals_df = murals_df[murals_df['Longitude'].notna()]
murals_df = murals_df[murals_df['Zip Code'].notna()]
murals_df
```

```python
murals_df.columns= murals_df.columns.str.lower()
murals_df
```

```python
murals_df['longitude'] = pd.to_numeric(murals_df['longitude'], errors='coerce')
landmarks_df['year installed'] = pd.to_numeric(landmarks_df['year installed'], errors='coerce')
```

```python
murals_df.dtypes
```

```python
landmarks_df.dtypes
```

```python
landmarks_df = landmarks_df[['artist', 'name', 'year installed', 'address', 'zip code', 'latitude', 'longitude', 'type', 'landmark designation date']]
landmarks_df.dtypes
```

```python
murals_df = murals_df[['artist', 'name', 'year installed', 'address', 'zip code', 'latitude', 'longitude', 'type', 'description of artwork']]
murals_df.dtypes
```

```python
chicago_df = landmarks_df.append(murals_df)
chicago_df
```

```python
chicago_df.columns = chicago_df.columns.str.replace(' ', '_')
chicago_df
```

```python
chicago_df['year_installed'].fillna(0, inplace=True)
chicago_df
```

```python
chicago_df['year_installed'] = chicago_df['year_installed'].astype(int)
chicago_df['zip_code'] = chicago_df['zip_code'].astype(str)

chicago_df
```

```python
chicago_df.columns
```

```python
chicago_df.dtypes
```

```python
chicago_df = chicago_df[chicago_df['zip_code'].notna()]
```

```python
chicago_df.to_csv("Resources/chicago_df.csv")
chicago_df
```

```python
chicago2_df=chicago_df.drop([102, 113, 115, 152])
chicago2_df
```

```python
chicago2_df.to_csv("Resources/chicago2_df.csv")
chicago2_df
```
